### SCATTER PLOT

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
# Make a scatter plot between sepal length and sepal width
```

```python
plt.scatter(x=iris['sepal_length'],y=iris['sepal_width'],color='green')
plt.xlabel('SL')
plt.ylabel('SW')
plt.title('Iris Flower',fontdict={'fontsize':14,'color':'darkred'})
plt.show()
```

```python
plt.scatter(x=iris['sepal_length'],y=iris['petal_length'],color='magenta',s=14,marker='^')
plt.xlabel('SL')
plt.ylabel('SW')
plt.title('Iris Flower',fontdict={'fontsize':14,'color':'darkred'})
plt.show()
```

```python
# Make a multiple scatter plot between sepal lenght sepal width, sepal length an pela length
```

```python
plt.scatter(x=iris['sepal_length'],y=iris['petal_length'],color='magenta',s=14,marker='^')
plt.scatter(x=iris['sepal_length'],y=iris['sepal_width'],s=14,color='darkgreen')
plt.xlabel('SL')
plt.ylabel('SW')
plt.title('Iris Flower',fontdict={'fontsize':14,'color':'darkred'})
plt.show()
```

```python
plt.figure(figsize=(7,3))

plt.suptitle('IRIS FLOWER',fontsize=16,color='darkred')

plt.subplot(1,2,1)
plt.scatter(x=iris['sepal_length'],y=iris['petal_length'],color='magenta',s=14,marker='^')
plt.xlabel('SEPAL LENGTH')
plt.ylabel('PETAL LENGTH')

plt.subplot(1,2,2)
plt.scatter(x=iris['sepal_length'],y=iris['sepal_width'],s=14,color='darkgreen')
plt.xlabel('SEPAL LENGTH')
plt.ylabel('SEPAL WIDTH')

plt.tight_layout() # its for fix the label
plt.show()
```

# BAR PLOT

```python
amount = [2500,3700,3000,4200,2200]
cust = ['Ajay','Suresh','Kamla','Neha','Vijay']
```

```python
# make a vertical barplot b/w amount and cust

plt.bar(x=cust,height=amount,width=0.50,color=['red','blue','green','m','c'])
plt.xlabel('customer name')
plt.ylabel('amount pending')
plt.yticks(range(1000,6000,1000))
plt.grid()
plt.show()
```

```python
plt.bar(x=cust,height=amount,width=0.50,color=['red','blue','green','m','c'])
for x,y in zip(range(5),amount):
    plt.text(x-0.20,y+50,str(y))
    
plt.xlabel('customer name')
plt.ylabel('amount pending')
plt.yticks(range(1000,6000,1000))
plt.xticks(rotation='vertical')
plt.grid()
plt.show()
```

```python
# make a horizontal bar plot b/w amount and cust
plt.barh(y=cust,width=amount,height=0.7,color=['red','blue','green','m','c'])
for x,y in zip(range(5),amount):
    plt.text(y+50,x-0.20,str(y))
    
plt.xticks(range(1000,6001,1000))
plt.grid()
plt.show()
```

# PIE PLOT

```python
countries = ['USA','France','Germany','India','Russia']
count = [63.76,34.54,45.11,57.43,60.54]
```

```python
# make a pie chart
plt.pie(x=count,labels=countries,colors=['red','blue','green','m','c'],
       startangle=90,counterclock=False,radius =1.2,autopct='%.0f%%',pctdistance=1.2,labeldistance=0.5,shadow=True,
       explode=[0,0,0,0.2,0])
plt.show()
```

```python
# make a doughnut chart
plt.pie(x=count,labels=countries,colors=['red','blue','green','m','c'],
       startangle=90,counterclock=False,radius =1.2,autopct='%.0f%%',pctdistance=1.2,labeldistance=0.7,shadow=True,
       explode=[0,0,0,0.2,0])
plt.pie(x=[9],colors=['w'],radius=0.6)
plt.show()
```

# Summary

### Summary
**Series Data**
linechart,areaplot

**numeric**
Distribution: Histogram, Distplot,Boxplot
    
**caregory**
Distribution: countplot(barchart),piechart
    
**numeric vs category**
stripplot,violinplot,swarmplot,boxplot

**numeric vs numeric**
scatter plot,pair plot

**category vs category**
crosstab,grouped barplot, stacked bar

**numeric vs numeric vs  hue=category**
scatter plot,pair plot

**numeric vs category vs hue = category**
stripplot,violinplot,swarmplot
