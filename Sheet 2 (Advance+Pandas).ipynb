### Concatination of series and array

```python
import pandas as pd
import numpy as np
```

```python
ser1 = pd.Series(np.random.randint(10,30,10))
ser2 = pd.Series(np.random.randint(10,30,10))
ser3 = pd.Series(np.random.randint(10,30,10))
```

```python
# Concatenate ser1 and ser2
# axis =0 (raw wise)
#axis =1 (column wise)

ans=pd.concat([ser1,ser2])
```

```python
ans[5]
```

```python
# concatenate ser1 and ser2 with hierarchial indexing
```

```python
 ans=pd.concat([ser1,ser2],keys = ['Data1','Data2'])
```

```python
ans['Data1'][5]
```

```python
ans['Data2'][5]
```

```python
# Concatenate the series with regular indexes (ignoring origional indexes)
```

```python
pd.concat([ser1,ser2],ignore_index =True)
```

```python
pd.concat([ser1,ser2],axis =1)
```

```python
pd.concat([ser1,ser2,ser3],axis =1)
```
